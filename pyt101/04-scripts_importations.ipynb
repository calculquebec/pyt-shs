### Scripts

Les programmes Python sont simplement un ensemble de commandes Python enregistrées dans un fichier. Pas besoin de compilation ! Pour le démontrer, écrivons notre tout premier programme !

Copiez le texte suivant dans un éditeur de texte et enregistrez-le sous le nom de votre choix (les fichiers Python ont généralement l’extension .py).

`print("je viens d'exécuter mon premier script python!!!")`

Nous pouvons maintenant exécuter ce programme de plusieurs façons. Si nous ouvrons un terminal dans le dossier où nous avons enregistré notre programme, nous pouvons utiliser la commande suivante pour l’exécuter :

`python mon_premier_script.py`

Les notebooks Python offrent aussi la possibilité d'exécuter du code Bash à même le notebook. Il s'uffit d'ajouter `!` avant la commande à exécuter.

`!python mon_premier_script.py`

### Importer des paquets Python

Python inclut de nombreuses fonctionnalités supplémentaires sous forme de paquets, mais tous ne sont pas chargés par défaut. Pour accéder à un paquet, nous devons l’importer. Certains de ses paquets sont généralistes, tandis que d'autres sont pour des utilisations très spécifiques. Voici quelques exemples de paquets populaires :

- pandas
- numpy
- matplotlib
- os
- seaborn
- scikit-learn

Pour rester dans l'esprit des scripts, nous allons tester l'importation du paquet `sys` qui permet entre autre d'accéder aux arguments de la ligne de commande.

```python
import sys
```

Pour tester les fonctionnalités de `sys`, copiez le texte suivant dans un éditeur de texte et enregistrez-le sous le nom de votre choix.

`import sys` 

`print('Nous avons écrit: ', sys.argv)`

Exécutez le script en ajoutant des mots ou des chiffres après le nom du script, par exemple :

`!python test_sys.py des choses pour tester ce paquet 1 2 3 4 5`

Sortie : 

`Nous avons écrit: ['test_sys.py', 'des', 'choses', 'pour', 'tester', 'ce', 'paquet', '1', '2', '3', '4', '5']`

La variable `sys.argv` est une liste qui contient les arguments passés au script via la ligne de commande.

Le paquet `sys` peut être très utile lorsqu'un script doit être utilisé plusieurs fois avec des fichiers d'entrée différents. Il permet de spécifier des variables à même la ligne de commande, sans avoir à changer le script.

### Quoi faire si le paquet est introuvable ?

Il peut arriver que vous tentiez d'importer un paquet et que vous obteniez un message d'erreur :

```python
import un_paquet_inexistant
```

La première chose à faire est de vérifier que vous n'avez pas fait une erreur de frappe dans le nom du paquet. Si l'erreur persiste, c'est que le paquet n'existe pas sur votre ordinateur et qu'il doit être téléchargé avant d'être importé. ### Anaconda

Si vous débutez en programmation, la façon la plus simple est de passer par Anaconda. Anaconda est une distribution libre et open source des langages de programmation Python et R qui vise à simplifier la gestion des paquets et de déploiement. Anaconda permet aussi de lancer des notebooks Jupyter, comme ceux utilisés dans cette formation. Si vous avez aimé la convivialité de Jupyter pendant la formation et que vous voulez poursuivre votre apprentissage de Python avec cet outil interactif, voici comment l'installer.

Pour installer Anaconda :
- Aller sur le site https://www.anaconda.com/download/success
- Télécharger la version pour votre système (Windows, macOS, Linux)
- Installer Anaconda (juste "Suivant", rien à configurer)

Pour installer un paquet PyPi :
- Cliquer sur Environments dans le menu de gauche
- Chercher le nom du paquet dans la barre de recherche
- Cocher le nom du paquet et cliquer sur Apply
- Si le paquet est introuvable dans la barre de recherche, c'est que son installation exige de passer par la ligne de commande. Les installations avancées seront présentées dans PYT102 Perfectionnement en programmation Python.

Pour lancer un Notebook :
- Ouvrir Anaconda Navigator
- Lancer JupyterLab via le Navigator avec le bouton "Launch". Cela ouvrira une interface web (dans votre navigateur par défaut). Ce n'est pas parce que ça utilise une connexion internet, mais seulement parce que les navigateurs sont parfaits pour afficher des interfaces interactives.
- Vous pouvez ensuite ouvrir des notebooks (extension .ipynb) à partir du navigateur de fichiers ou en créer un nouveau.
